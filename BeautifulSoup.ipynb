```python
import bs4
import json
import pandas as pd
```

```python
#I downloaded the url "http://www.burjeel.com/abu-dhabi/doctors/" in html format
x = open("Doctors.html", 'r').read()
```

```python
#Converted HTML page to a soup objet

from bs4 import BeautifulSoup
soup = BeautifulSoup(x, 'html.parser')
print(soup.prettify())
```

```python
doctor_name = []
for i in soup.find_all(attrs = {"class":"col-lg-4 col-md-4 col-sm-6 col-xs-12 list_wrapper innercommon"}):
    doctor_name.append(i.a.h4.string)
```

```python
Designation = []
for i in soup.find_all('h5'):
    Designation.append(i.string)
```

```python
x = []
with open('Data.jl','a') as file:
    for (a,b) in zip(doctor_name, Designation):
        x.append({"Doctor":a, "Designation":b})
    json.dump(x, file)
```
